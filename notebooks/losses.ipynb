```python
import numpy as np
import matplotlib.pyplot as plt

# Define the error values
x = np.linspace(-4, 4, 400)
delta = 1.5

# MSE Loss
mse_loss = x**2

# Huber Loss
huber_loss = np.where(np.abs(x) <= delta, 0.5 * x**2, delta * (np.abs(x) - 0.5 * delta))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(x, mse_loss, label='MSE Loss', color='blue')
plt.plot(x, huber_loss, label='Huber Loss', color='red', linestyle='dashed')
plt.title('Comparison of MSE Loss and Huber Loss')
plt.xlabel('Error (x)')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def asymmetric_mse_loss(y_true, y_pred):
    """Compute the asymmetric MSE loss."""
    residual = y_true - y_pred
    alpha = 2  # Penalty multiplier for underestimations
    loss = np.where(residual < 0, alpha * residual**2, residual**2)
    return loss

# Generate residuals
residuals = np.linspace(-3, 3, 400)
y_true = 0  # Set true value to 0 for visualization purposes
losses = asymmetric_mse_loss(y_true, y_true - residuals)

# Plotting the loss function
plt.figure(figsize=(8, 5))
plt.plot(residuals, losses, label='Asymmetric MSE Loss', color='blue')
plt.title('Asymmetric MSE Loss Function')
plt.xlabel('Residual (y_true - y_pred)')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def asymmetric_mse_loss(y_true, y_pred, alpha):
    """Compute the asymmetric MSE loss with a given penalty multiplier for underestimations."""
    residual = y_true - y_pred
    loss = np.where(residual < 0, alpha * residual**2, residual**2)
    return loss

# Generate residuals
residuals = np.linspace(-3, 3, 400)
y_true = 0  # Using true value 0 for visualization

# Different alpha values
alphas = [1, 2, 3, 5]
colors = ['blue', 'green', 'red', 'purple']
labels = [f'Asymmetric MSE Loss (α={alpha})' for alpha in alphas]

# Plotting the loss functions
plt.figure(figsize=(10, 6))
for alpha, color, label in zip(alphas, colors, labels):
    losses = asymmetric_mse_loss(y_true, y_true - residuals, alpha)
    plt.plot(residuals, losses, label=label, color=color)

plt.title('Comparison of Asymmetric MSE Loss Functions with Different Penalties')
plt.xlabel('Residual (y_true - y_pred)')
plt.ylabel('Loss')
plt.grid(True)
plt.legend()
plt.show()
```
